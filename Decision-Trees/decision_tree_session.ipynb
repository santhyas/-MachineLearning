```python
import pandas as pd
```

```python
df = pd.read_csv('loan_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
purpose = pd.get_dummies(df['purpose'], drop_first=True)
```

```python
df.drop('purpose', axis =1, inplace=True)
```

```python
df = pd.concat([purpose, df], axis=1)
```

```python
df.head()
```

```python
df.corr()
```

```python
import seaborn as sns
```

```python
sns.heatmap(df.corr(),cmap='coolwarm')
```

```python
df.drop(['home_improvement','delinq.2yrs'],axis = 1, inplace = True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(df.drop('not.fully.paid', axis =1))
scaled_array = ss.transform(df.drop('not.fully.paid', axis =1))
```

```python
X  = pd.DataFrame(data = scaled_array, columns = df.columns[:-1])
y = df['not.fully.paid']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
ypred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, ypred)
```
